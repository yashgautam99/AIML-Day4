```python
import numpy as np
import matplotlib.pyplot as plt
# create a ara of 10x10 kms and deploy the
# 50 nodes in it
num_of_sensors = 100
np.random.seed(5)
# generate the data
#data = np.random.randint(0,100,size=(num_of_sensors,2))
data = 0+(100-0)*np.random.random((num_of_sensors,2))
# display the sensors nodes
plt.figure(1)
plt.plot(data[:,0],data[:,1],'ro')
#s - square , d - diamond, o-circle,*-star
# data.shape will return num of rows and col of data
# .shape[0] will rows and .shape[1] cols
for i in range(data.shape[0]):
    dx = data[i,0]
    dy = data[i,1]
    plt.text(dx-1,dy+2,str(i+1))
```

```python
# pre allocate the size for the distance matrix
dist_matrix = np.zeros((data.shape[0],data.shape[0]))

# run a logic to calculate the distance of each node wrt
# each node
for i in range(data.shape[0]):
  for j in range(data.shape[0]):
    dist = np.sqrt((data[i,0]-data[j,0])**2 + (data[i,1]-data[j,1])**2)
    dist_matrix[i,j]=dist
# print the dist_matrix
dist_matrix.astype(int)
```

```python
# create a neighbor matrix
neigh_matrix = np.zeros((data.shape[0],data.shape[0]))
SR = 40 # define sensing range
for i in range(data.shape[0]):
  for j in range(data.shape[0]):
    # chk only non- diagonal elements
    if(i !=j):
      if(dist_matrix[i,j] <= SR):
        neigh_matrix[i,j]=1
# print the neighbor matrix
neigh_matrix
```

```python
# ask the user to enter the node number
# print its nighbors and visualize
node = int(input("Enter the Node number: "))
node = node-1
temp_neigh = neigh_matrix[node,:]
tNeigh=[]# empty list
for i in range(len(temp_neigh)):
  if(temp_neigh[i]==1):
    tNeigh.append(i)
print("Node Number",node)
print("Neighboring nodes are: ",tNeigh)

plt.figure(1)
plt.plot(data[:,0],data[:,1],'ro')
#s - square , d - diamond, o-circle,*-star
# data.shape will return num of rows and col of data
# .shape[0] will rows and .shape[1] cols
for i in range(data.shape[0]):
    dx = data[i,0]
    dy = data[i,1]
    plt.text(dx-1,dy+2,str(i+1))
    plt.plot(data[node,0],data[node,1],'bd')
for i in range(len(tNeigh)):
  Xtemp = [data[node,0],data[tNeigh[i],0]]
  Ytemp = [data[node,1],data[tNeigh[i],1]]
  plt.plot(Xtemp,Ytemp,'r',linewidth=2)
  plt.plot(data[tNeigh[i],0],data[tNeigh[i],1],'gs')
```
